```python
# Import necessary modules
from IPython.display import display, HTML
import ipywidgets as widgets
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.llms import Ollama

# Initialize the Ollama model
llm = Ollama(model="llama2")

# Initialize the conversation history
history = [("system", "You are a helpful assistant. Please respond to the questions")]

# Define a prompt template for the chatbot
def build_prompt(history):
    return ChatPromptTemplate.from_messages(history)

# Function to handle input and update conversation
def handle_input(change):
    input_text = change['new']
    if input_text:
        history.append(("user", f"Question: {input_text}"))
        prompt = build_prompt(history)
        chain = prompt | llm
        response = chain.invoke({"question": input_text})
        history.append(("assistant", response))
        
        conversation = ""
        for role, text in history:
            if role == "user":
                conversation += f"<b>You:</b> {text.split('Question: ')[-1]}<br>"
            elif role == "assistant":
                conversation += f"<b>Assistant:</b> {text}<br>"
        
        conversation_box.value = conversation
        input_box.value = ""

# Create the input box and conversation display
input_box = widgets.Text(placeholder="Ask your question!")
input_box.observe(handle_input, names='value', type='change')
conversation_box = widgets.HTML(value="", placeholder="Conversation will appear here...", layout=widgets.Layout(height='400px', overflow_y='scroll', border='1px solid #ccc'))

# Display the chat interface
display(HTML("<h1>Langchain Chatbot With LLAMA2 model</h1>"))
display(conversation_box)
display(input_box)

# Initialize the conversation display
conversation_box.value = "<b>Assistant:</b>I am your a helpful assistant. How may I be of assistance?<br>"
```
